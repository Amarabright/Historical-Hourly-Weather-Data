## importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

%matplotlib inline
```

### importing datasets

```python
pressure = pd.read_csv("pressure.csv")
pressure.head(10)
```

```python
humidity = pd.read_csv("humidity.csv")
humidity.head(10)
```

```python
temperature = pd.read_csv("temperature.csv")
temperature.head(10)
```

```python
weather = pd.read_csv("weather_description.csv")
weather.head(10)
```

```python
wind_dir = pd.read_csv("wind_direction.csv")
wind_dir.head(10)
```

```python
wind_speed = pd.read_csv("wind_speed.csv")
wind_speed.head(10)
```

### creating new dataset by city

```python
# Extracting data for Chicago
chicago = (pressure[['datetime', 'Chicago']].merge(humidity[['datetime', 'Chicago']], on = 'datetime')
           .merge(temperature[['datetime', 'Chicago']], on = 'datetime').merge(weather[['datetime', 'Chicago']], on = 'datetime').merge(wind_dir[['datetime', 'Chicago']], on = 'datetime')
           .merge(wind_speed[['datetime', 'Chicago']], on = 'datetime'))
```

```python
chicago.columns = ['datetime', 'pressure', 'humidity', 'temperature', 'weather', 'wind direction', 'wind speed']
chicago.head()
```

```python
# Extracting data for Nashville
nashville = (pressure[['datetime', 'Nashville']].merge(humidity[['datetime', 'Nashville']], on = 'datetime')
            .merge(temperature[['datetime', 'Nashville']], on = 'datetime').merge(weather[['datetime', 'Nashville']], on = 'datetime')
            .merge(wind_dir[['datetime', 'Nashville']], on = 'datetime').merge(wind_speed[['datetime', 'Nashville']], on = 'datetime'))
```

```python
nashville.columns = ['datetime', 'pressure', 'humidity', 'temperature', 'weather', 'wind direction', 'wind speed']
nashville.head()
```

```python
# Extracting data for Saint Louis
sl = (pressure[['datetime', 'Saint Louis']].merge(humidity[['datetime', 'Saint Louis']], on = 'datetime')
            .merge(temperature[['datetime', 'Saint Louis']], on = 'datetime').merge(weather[['datetime', 'Saint Louis']], on = 'datetime')
            .merge(wind_dir[['datetime', 'Saint Louis']], on = 'datetime').merge(wind_speed[['datetime', 'Saint Louis']], on = 'datetime'))
```

```python
sl.columns = ['datetime', 'pressure', 'humidity', 'temperature', 'weather', 'wind direction', 'wind speed']
sl.head()
```

## Performing EDA on each city

### Chicago:

### Correlation plot 

```python
plt.figure(figsize = (12, 10))
sns.heatmap(data = chicago.corr(), annot = True)
plt.show()
```

### distribution plot for numerical features

```python
# plot distributions of each feature
plt.figure(figsize = (15,15))
# Pressure distribution
fig = plt.subplot(321)
sns.distplot(chicago['pressure'])
plt.title('Pressure distribution for Chicago')

# Humidity distribution
plt.subplot(322)
sns.distplot(chicago['humidity'])
plt.title("Humidity distribution for Chicago")

# Temperature distribution
plt.subplot(323)
sns.distplot(chicago['temperature'])
plt.title("Temperature distribution for Chicago")

# Wind direction distribution
plt.subplot(324)
sns.distplot(chicago['wind direction'])
plt.title("Wind direction distribution for Chicago")

plt.subplot(325)
sns.distplot(chicago['wind speed'])
plt.title("Wind speed distribution for Chicago")
```

### distribution plot of numerical features average per year

```python
plt.figure(figsize = (15,15))
plt.subplot(321)
chicago.groupby('Year')['pressure'].mean().plot.barh(color = 'purple')
plt.xlabel('Pressure')
plt.title('Mean Pressure per Year')

plt.subplot(322)
chicago.groupby('Year')['temperature'].mean().plot.barh(color = 'red')
plt.xlabel('Temperature')
plt.title("Mean Temperature per Year")

plt.subplot(323)
chicago.groupby('Year')['humidity'].mean().plot.barh(color = 'blue')
plt.xlabel('Humidity')
plt.title('Mean Humidity per Year')

plt.subplot(324)
chicago.groupby('Year')['wind direction'].mean().plot.barh(color = 'yellow')
plt.xlabel('Humidity')
plt.title('Mean wind direction per Year')

plt.subplot(325)
chicago.groupby('Year')['wind speed'].mean().plot.barh(color = 'magenta')
plt.xlabel('Humidity')
plt.title('Mean Humidity per Year')
```

### distribution plot of numerical features per month 

```python
plt.figure(figsize = (15,15))
plt.subplot(321)
chicago.groupby('Month')['pressure'].mean().plot.bar(color = 'purple')
plt.xlabel('Pressure')
plt.title('Mean Pressure per Month')

plt.subplot(322)
chicago.groupby('Month')['temperature'].mean().plot.bar(color = 'red')
plt.xlabel('Temperature')
plt.title("Mean Temperature per Month")

plt.subplot(323)
chicago.groupby('Month')['humidity'].mean().plot.bar(color = 'blue')
plt.xlabel('Humidity')
plt.title('Mean Humidity per Month')

plt.subplot(324)
chicago.groupby('Month')['wind direction'].mean().plot.bar(color = 'yellow')
plt.xlabel('Humidity')
plt.title('Mean wind direction per Month')

plt.subplot(325)
chicago.groupby('Month')['wind speed'].mean().plot.bar(color = 'magenta')
plt.xlabel('Humidity')
plt.title('Mean Humidity per Month')
```

### Plotting weather per month
