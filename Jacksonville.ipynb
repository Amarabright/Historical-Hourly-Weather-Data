### Extracting Features of Jacksonville

```python
jacksonville = (humidity[['datetime', 'Jacksonville']].merge(pressure[['datetime', 'Jacksonville']], on = 'datetime')
                .merge(temp[['datetime', 'Jacksonville']], on = 'datetime').merge(weather_desc[['datetime', 'Jacksonville']], on = 'datetime')
                .merge(wind_dir[['datetime', 'Jacksonville']], on = 'datetime').merge(wind_speed[['datetime', 'Jacksonville']], on = 'datetime'))
jacksonville.columns = ['datetime', 'humidity', 'pressure', 'temperature', 'weather_description', 'wind_direction', 'wind_speed']
jacksonville.head()
```

### Checking for missing values

```python
jacksonville.isnull().sum()
```

```python
#Since 1st row is full of null values, we can drop it directly
jacksonville = jacksonville.drop(0, axis=0)
jacksonville.reset_index(drop=True, inplace=True)
jacksonville.head()
```

```python
#Checking mean of humidity as a means for replacing NaN values
np.around(jacksonville['humidity'].mean())
```

```python
jacksonville['humidity'].unique()
```

```python
#plotting a box plot to figure out if outliers are present in the dataset
px.box(jacksonville['humidity'])
```

```python
#Checking mean of pressure as a means for replacing NaN values
jacksonville['pressure'].mode()
```

```python
jacksonville['pressure'].unique()
```

```python
#plotting a box plot to figure out if outliers are present in the dataset
px.box(jacksonville['pressure'])
```

```python
#Filling in NaN values for humidity and pressure using rounded mean and mode
jacksonville['humidity'].fillna(np.around(jacksonville['humidity'].mean()), inplace=True)
jacksonville['pressure'].fillna(1017, inplace=True)
jacksonville.isnull().sum()
```

```python
#Filling in rest of NaN values
jacksonville.fillna(method='ffill', inplace=True)
jacksonville.isnull().sum()
```

```python
#Converting datetime column to seperate columns of Year, month, day and hours
date = pd.DatetimeIndex(jacksonville['datetime'])
jacksonville['Year'] = date.year
jacksonville['Month'] = date.month
jacksonville['Day'] = date.day
jacksonville['Hours'] = date.hour
jacksonville.head()
```

```python
#Removing datetime column as we have seperated it
jacksonville.drop(['datetime'], axis=1, inplace=True)
jacksonville.head()
```

### EDA on Jacksonville
